# APIM 경유 Azure OpenAI 호출 테스트

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(os.path.join(os.path.dirname(__file__) if '__file__' in dir() else '../..', '.env'))
load_dotenv('../../.env')  # notebooks are 2 levels deep

ENDPOINT = os.environ["ENDPOINT"]
API_KEY = os.environ["API_KEY"]

client = OpenAI(
    api_key="placeholder",
    base_url=ENDPOINT,
    default_headers={"api-key": API_KEY},
)

response = client.chat.completions.create(
    model="gpt-5.4",
    messages=[{"role": "user", "content": "안녕 Azure에 대해 설명해줘"}],
    max_completion_tokens=300,
)

print(f"✅ APIM 정상 작동 확인!")
print(f"모델: {response.model}")
print(f"응답: {response.choices[0].message.content}")
```
